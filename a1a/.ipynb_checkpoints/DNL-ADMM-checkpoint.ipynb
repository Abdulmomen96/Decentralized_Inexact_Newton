```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
import os

import scipy
import networkx as nx

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from numpy import linalg as LA
```

```python
seed = 42
no_users = 16
lambda_logistic = 1e-3
```

```python
def random_split(X, y, n, seed):
    """Equally split data between n agents"""
    rng = np.random.default_rng(seed)
    perm = rng.permutation(y.size)
    X_split = np.array_split(X[perm], n)  #np.stack to keep as a np array
    y_split = np.array_split(y[perm], n)
    return X_split, y_split
```

```python
def loss(w, A, b, l2):
    z = b * np.dot(A, w)
    tmp = np.minimum(z, 0)
    loss = np.log((np.exp(tmp) + np.exp(tmp - z)) / np.exp(tmp))
    loss_sum = np.sum(loss) / len(b)
    reg = (np.linalg.norm(w) ** 2) * l2 / 2
    return loss_sum + reg
```

```python
def gradient(w, A, b, l2):
    m = A.shape[0]
    z = b * np.dot(A, w)
    tmp0 = np.minimum(z, 0)
    tmp1 = np.exp(tmp0 - z) / ((np.exp(tmp0) + np.exp(tmp0 - z)))
    tmp2 = - tmp1 * b
    res = np.dot(A.T, tmp2) / m + l2 * w
    return res
```

```python
def hessian(w, A, b, l2):
    Aw = A @ w
    activation = scipy.special.expit(Aw)
    weights = activation * (1-activation)
    A_weighted = np.multiply(A.T, weights)
    return A_weighted@A/A.shape[0] + l2*np.eye(A.shape[1])
```

```python
def generate_graph(n, seed):
    """Generate a random connected graph"""
    while True:
        g = nx.generators.random_graphs.binomial_graph(n, 0.4, seed = seed) 
        # g = nx.random_geometric_graph(n, 0.4, seed = seed)
        #g = nx.grid_2d_graph(int(n/5), int(n/5))
        if nx.algorithms.components.is_connected(g):
            return g
```

```python
G = generate_graph(no_users, seed)
adjacency_matrix = nx.linalg.graphmatrix.adjacency_matrix(G)
print(G.number_of_edges())
print(G.number_of_nodes())
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
X = np.load('X.npy')
y = np.load('y.npy')

num_feature = X.shape[1] #+ 1 #+1 for bias
num_samples = X.shape[0]
```

```python
X, y = random_split(X, y, no_users, seed)

grad = [np.zeros(num_feature) for _ in range(no_users)] # old grads
Hess = [np.zeros([num_feature, num_feature]) for _ in range(no_users)] # old hessians
```

```python
# Optimal objective function, i.e., f(x*)
obj0 = 0.32792319329870895 # a1a dataset
#obj0 = 0.3333472060757055 # a9a dataset
# obj0 = 0.1834730582750861 # w8a dataset
# obj0 = 0.4076194423164931 # phishing dataset
```

```python
n=num_feature
P = adjacency_matrix.toarray()
L_neg=np.diag(np.sum(P,0))-P
L_pos=np.abs(L_neg)
alfa_0=np.zeros(n*no_users)
x_0=np.zeros(n*no_users)
```

```python
def Total_hessian(Hess, n, no_users):
    H = np.zeros([n*no_users,n*no_users])
    for i in range(no_users):
        H[i*n:(i+1)*n,i*n:(i+1)*n] = Hess[i]
    return H
```

```python
def Total_grad(grad, n, no_users):
    g = np.zeros([n*no_users])
    for i in range(no_users):
        g[i*n:(i+1)*n] = grad[i]
    return g
```

```python
def ADMM_Consensus_LS_xalpha_updates(MtMplus2cW,Mty,cL_pos,cL_neg,alfa,x):
    x=np.matmul(np.linalg.inv(MtMplus2cW),Mty+np.matmul(cL_pos,x)-alfa)
    alfa=alfa+np.matmul(cL_neg,x)
    return x, alfa
```

```python
class FactorScheduler:
    def __init__(self, factor=1, stop_factor_lr=1e-7, base_lr=0.1):
        self.factor = factor
        self.stop_factor_lr = stop_factor_lr
        self.base_lr = base_lr

    def __call__(self, num_update):
        self.base_lr = max(self.stop_factor_lr, self.base_lr * self.factor)
        return self.base_lr
```

```python
#rho = 0.005
d=x_0
theta =x_0
alfa=alfa_0
P=np.kron(P,np.eye(n))
L_neg=np.kron(L_neg,np.eye(n))
```

```python
class FactorScheduler:
    def __init__(self, factor=1, stop_factor_lr=1e-7, base_lr=0.1):
        self.factor = factor
        self.stop_factor_lr = stop_factor_lr
        self.base_lr = base_lr

    def __call__(self, num_update):
        self.base_lr = max(self.stop_factor_lr, self.base_lr * self.factor)
        return self.base_lr
```

```python
n_iters = 200
# increasing base_lr we avoid the issue of flat curve
#factor=0.95, stop_factor_lr=1e-4, base_lr=0.1, no_loc=5
scheduler = FactorScheduler(factor=0.95, stop_factor_lr=1e-3, base_lr=10.0)
rho = [scheduler(t) for t in range(n_iters)]
losses_dnl = []
op_gap_dnl = np.zeros(shape=[n_iters])
theta_temp = [np.zeros(num_feature) for _ in range(no_users)]
no_loc = 1

for k in range(n_iters):
    
    cL_neg=rho[k]*L_neg
    cL_pos=np.abs(cL_neg)
    
    for i in range(no_users):
        grad[i] = gradient(theta_temp[i], X[i], y[i], lambda_logistic)
        Hess[i] = hessian(theta_temp[i].ravel(), X[i], y[i], lambda_logistic)
    
    
    MtMplus2cW=Total_hessian(Hess, n, no_users) + 2*rho[k]*np.diag(np.sum(P,0))
    Mty = Total_grad(grad, n, no_users)
    
    for i in range(no_loc):
        [d,alfa]=ADMM_Consensus_LS_xalpha_updates(MtMplus2cW,Mty,cL_pos,cL_neg,alfa,d)
    
    theta = theta - d
    
    theta_temp = [np.zeros(num_feature) for _ in range(no_users)]
    for j in range(no_users):
        theta_temp[j] = theta[j*n:(j+1)*n]
        
    # Performance Check
    
    theta_avg = 1/no_users*sum(theta_temp)
    
    loss_dnl = np.mean([loss(theta_avg, X[i], y[i], lambda_logistic) for i in range(no_users)])
    losses_dnl.append(loss_dnl)
    
    op_gap_dnl[k] = np.abs(losses_dnl[-1] - obj0)
```
