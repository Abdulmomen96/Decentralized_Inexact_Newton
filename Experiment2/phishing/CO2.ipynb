```python
import eco2ai
import numpy as np
import pandas as pd
tracker = eco2ai.Tracker(project_name="DNL", experiment_description="training DNL")

tracker.start()

for i in range(1000):
    temp_mat = np.random.rand(100, 100)
    inv_mat = np.linalg.inv(temp_mat)
    print(np.linalg.norm(temp_mat))
    print(np.linalg.norm(inv_mat))
tracker.stop()
```
