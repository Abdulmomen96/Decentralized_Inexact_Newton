```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
re_dnl = np.load('re_dnl.npy').ravel()
re_dsgd = np.load('re_dsgd.npy').ravel()
re_newton_tracking = np.load('re_newton_tracking.npy').ravel()
re_nn = np.load('re_nn.npy').ravel()
```

```python
plt.semilogy(re_dsgd, label='DGD') # Descentralized Gradient Descent
plt.semilogy(re_nn, label='NN') # Network Newton
plt.semilogy(re_newton_tracking, label='NT') # Newton Tracking
plt.semilogy(re_dnl, label='DNL') # Ours


plt.xlabel('Number of Iterations')
plt.ylabel('Optimality Gap')
legend = plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.grid(True)
#plt.ylim([10**(-5),0.5])
plt.savefig("Optimality Gap.pdf")
```
